```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
# from sklearn.preprocessing import StandardScaler, MinMaxScaler # Commented out as data appears pre-scaled

# Modified to read the uploaded file. Adjust filename if needed.
df = pd.read_csv("Automobile_standard_scaled_0131.csv")

print("Initial Data Info:")
print(df.info())

print("\nMissing values per column:\n", df.isnull().sum())

# Assuming 'horsepower' was dropped or not present in the scaled data, keep this check.
if "horsepower" in df.columns:
    df.drop("horsepower", axis=1, inplace=True)

num_cols = df.select_dtypes(include=["int64", "float64"]).columns

# Imputation might still be necessary if there are missing values in the scaled data
imputer = SimpleImputer(strategy="median")
df[num_cols] = imputer.fit_transform(df[num_cols])

# Removed scaling steps as data appears pre-scaled
# scaler_std = StandardScaler()
# df_std = df.copy()
# df_std[num_cols] = scaler_std.fit_transform(df_std[num_cols])

# scaler_mm = MinMaxScaler()
# df_mm = df.copy()
# df_mm[num_cols] = scaler_mm.fit_transform(df_mm[num_cols])

# Save the processed data (which is now the loaded and potentially imputed scaled data)
df.to_csv("Automobile_Processed.csv", index=False)
# Removed saving of separate scaled files as input was already scaled
# df_std.to_csv("Automobile_Standardized.csv", index=False)
# df_mm.to_csv("Automobile_MinMax.csv", index=False)


print("\n✅ Preprocessing Completed!")
print("Saved file: Automobile_Processed.csv") # Updated saved filename

print("\nReasoning:")
print("👉 Data was loaded from the uploaded standardized file.")
print("👉 Imputation was performed to handle any missing values.")
print("👉 Scaling steps were skipped as the input data appears to be pre-scaled.")
```
